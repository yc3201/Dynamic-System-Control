### 1.18

### i) Original ODE system

```python
import numpy as np
from scipy.integrate import odeint
import matplotlib.pyplot as plt
```

```python
def model_original(z,t):
    dxdt = z[0]**2+z[1]**2+z[1]*np.cos(z[0])
    dydt = (1+z[0])*z[0]+(1+z[1])*z[1]+z[0]*np.sin(z[1])
    dzdt = [dxdt,dydt]
    return dzdt

# initial condition
z0 = [0.0001,0.0001]

# time points
t = np.linspace(0,5,100)

# solve ODE
z = odeint(model_original,z0,t)

# plot results
plt.plot(t,z[:,0],'b-',label=r'$z_1$')
plt.plot(t,z[:,1],'r--',label=r'$z_2$')
plt.ylabel('response')
plt.xlabel('time')
plt.title('Nonlinear System with IV[0.0001,0.0001]',fontsize=20 )
plt.legend(loc='best')
plt.grid()
```

```python
### Change for different initial values
IV = [[0.0001,0.0001],[-0.0001,-0.0001],[-0.0001, 0.0001],[0.0001, -0.0001]]
color_list = ['orange','turquoise','lightpink','mediumslateblue']
for z0 in IV:
    # solve ODE
    z = odeint(model_original,z0,t)
    # plot results
    plt.plot(t,z[:,0],'b-',label=r'$z_1$'+str(z0))
    plt.plot(t,z[:,1],'r--',label=r'$z_2$'+ str(z0))
    plt.ylabel('response')
    plt.xlabel('time')
    plt.title('Nonlinear System with IV' +str(z0),fontsize=20 )
    plt.legend(loc='best')
    plt.show()
```

### ii) Linear System

```python
def model(z,t):
    dxdt = z[1]
    dydt = z[0] + z[1]
    dzdt = [dxdt,dydt]
    return dzdt

# initial condition
z0 = [0.0001,0.0001]

# time points
t = np.linspace(0,5,100)

# solve ODE
z = odeint(model,z0,t)

# plot results
plt.plot(t,z[:,0],'b-',label=r'$z_1$')
plt.plot(t,z[:,1],'r--',label=r'$z_2$')
plt.ylabel('response')
plt.xlabel('time')
plt.legend(loc='best')
plt.grid()
```

```python
### Change for different initial values
IV = [[0.0001,0.0001],[-0.0001,-0.0001],[-0.0001, 0.0001],[0.0001, -0.0001]]
color_list = ['orange','turquoise','lightpink','mediumslateblue']
for z0 in IV:
    # solve ODE
    z = odeint(model,z0,t)
    # plot results
    plt.plot(t,z[:,0],'b-',label=r'$z_1$'+str(z0))
    plt.plot(t,z[:,1],'r--',label=r'$z_2$'+ str(z0))
    plt.ylabel('response')
    plt.xlabel('time')
    plt.title('Linear System with IV' +str(z0),fontsize=20 )
    plt.legend(loc='best')
    plt.show()
```

Changing initial conditions in the vicinty of the origin of the above system of equitions and its lineariztion will alter the solution in a similar way with different magnitudes.

### 1.33

### i) Nonlinear Model

```python
def pendulum_nl(z,t):
    dxdt = z[1]
    dydt = -10*np.sin(z[0])
    dzdt = [dxdt,dydt]
    return dzdt

# initial condition
z0 = [np.pi/18,0]

# time points
t = np.linspace(0,10,100)

# solve ODE
z = odeint(pendulum_nl,z0,t)

# plot results
plt.plot(t,z[:,0],'b-',label=r'$x_1$')
plt.plot(t,z[:,1],'r--',label=r'$x_2$')
plt.ylabel('response')
plt.xlabel('time')
plt.title('Nonlinear Simple Pendulum with IV'+r'$[\pi/18,0]$',fontsize=20 )
plt.legend(loc='best')
plt.grid()
```

```python
### Change for different initial values
IV = [[np.pi/12,0],[np.pi/6,0],[np.pi/3, 0]]
for z0 in IV:
    # solve ODE
    z = odeint(pendulum_nl,z0,t)
    # plot results
    plt.plot(t,z[:,0],'b-',label=r'$x_1$'+str(z0))
    plt.plot(t,z[:,1],'r--',label=r'$x_2$'+ str(z0))
    plt.ylabel('response')
    plt.xlabel('time')
    plt.title('Nonlinear Simple Pendulum with IV' +str(z0),fontsize=20 )
    plt.legend(loc='best')
    plt.show()
```

 ### ii) Linear System

```python
def pendulum_l(z,t):
    dxdt = z[1]
    dydt = -10*z[0]
    dzdt = [dxdt,dydt]
    return dzdt

# initial condition
z0 = [np.pi/18,0]

# time points
t = np.linspace(0,10,100)

# solve ODE
z = odeint(pendulum_l,z0,t)

# plot results
plt.plot(t,z[:,0],'b-',label=r'$x_1$')
plt.plot(t,z[:,1],'r--',label=r'$x_2$')
plt.ylabel('response')
plt.xlabel('time')
plt.title('Linear Simple Pendulum with IV'+r'$[\pi/18,0]$',fontsize=20 )
plt.legend(loc='best')
plt.grid()
```

```python
### Change for different initial values
IV = [[np.pi/12,0],[np.pi/6,0],[np.pi/3, 0]]
for z0 in IV:
    # solve ODE
    z = odeint(pendulum_l,z0,t)
    # plot results
    plt.plot(t,z[:,0],'b-',label=r'$x_1$'+str(z0))
    plt.plot(t,z[:,1],'r--',label=r'$x_2$'+ str(z0))
    plt.ylabel('response')
    plt.xlabel('time')
    plt.title('Linear Simple Pendulum with IV' +str(z0),fontsize=20 )
    plt.legend(loc='best')
    plt.show()
```

Within the system, changing initial conditions will clearly change the plots of $x_1$ and $x_2$. Between two system, given the same initial condition, there is no differnce between the plots of $x_1$ and $x_2$ in two systems. 

### 1.12

### i) $g(x) = k(1+a^2x2)x$

```python
def dF(p, q):
    a = 0.05
    k = .6
    return q, -k*(1+ a**2*p**2)*p

X, Y = np.meshgrid(np.linspace(-4.0, 4.0, 30), np.linspace(-4.0, 4.0, 30))
dp, dq = dF(X, Y)
U, V = dp, dq

plt.streamplot(X, Y, U, V, color='r', linewidth=0.5, density=1.6)
plt.axis('square')
plt.axis([-4, 4, -4, 4])
plt.xlabel('$x_1$')
plt.ylabel('$x_2$')
plt.show()
```

### ii) $g(x) = kx$

```python
def dF(p, q):
    k = .6
    return q, -k*p

X, Y = np.meshgrid(np.linspace(-4.0, 4.0, 30), np.linspace(-4.0, 4.0, 30))
dp, dq = dF(X, Y)
U, V = dp, dq

plt.streamplot(X, Y, U, V, color='r', linewidth=0.5, density=1.6)
plt.axis('square')
plt.axis([-4, 4, -4, 4])
plt.xlabel('$x_1$')
plt.ylabel('$x_2$')
plt.show()
```
